```python
import pandas as pd
import glob
keyword = '游戏类'
path = './' + keyword + '/wordtracker' # use your path
all_files = glob.glob(path + "/*.csv")
print(all_files)
```

```python
li = []
for filename in all_files:
    f=open(filename, 'rb')
    df = pd.read_csv(f, encoding='utf8', header=0)
    li.append(df)
frame = pd.concat(li, axis=0, ignore_index=True).set_index('Keyword')
```

```python
youtube_vol = frame[['Volume','Competition','IAAT','KEI']]
youtube_vol = youtube_vol.rename(columns={"Volume": "youtubeVol"})
youtube_vol['Competition'] = youtube_vol['Competition'].str.replace("-", "0").astype(float)
youtube_vol['IAAT'] = youtube_vol['IAAT'].str.replace("-", "0").astype(float)
```

```python
joined = youtube_vol.drop_duplicates()
joined
```

```python
joined.to_csv('./'+ keyword + '/joined_wordtracker.csv')
```

# weighted scores

```python
import numpy as np
```

```python
joined['new_KEI']=(np.log(joined.youtubeVol)/joined.Competition).replace([np.inf, -np.inf], np.nan)
joined['vol_comp']= joined.youtubeVol*joined.Competition.replace([np.inf, -np.inf], np.nan)
```

```python
sorted_new_KEI= joined.sort_values(by='new_KEI', ascending=False).head(10)
sorted_vol_comp= joined.sort_values(by='vol_comp', ascending=False).head(10)
```
